```python
def findseq(instr):
    temp = re.findall(r'\d+', instr)
    nums = []
    
    if not temp:
        return False
        
    for i in temp[0]:
        nums.append(int(i))
    if len(nums) < 3:
        return False
    i = 0
    matches = 0
    while i < 2:
        if nums[i] + 1 == nums[i+1]:
            matches = matches + 1
        i = i+1
        
    if matches >= 2:
        return True
    
    i = 1
    matches = 0
    while i < 3:
        if nums[-i] + 1 == nums[-i-1]:
            matches = matches + 1
        i = i+1
    
    if matches >= 2:
        return True
    
    return False
```

```python
viz2list = []
for i in Viz2:
    viz2list.append(findseq(i))
```

```python
def getvowels(lists):
    vowels = ['a','e','i','o','u','y']
    letters = []
    for i in lists:
        for i2 in i:
            if i2 in vowels:
                letters.append(i2)
    return letters
```
